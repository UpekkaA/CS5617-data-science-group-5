```python
#Wage deviation between male and female province wise
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('resources/Inequality_among_gender.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    i=0
    for row in plots:
        if i==0:
            year = list(map(int, (row[4:11])))
        #print(row[4:9])
        Gender=row[3]
        province=row[0]
        sector=row[2]
        if Gender=='Male':
            male = list(map(int, (row[4:11])))
        if Gender == 'Female':
            female = list(map(int, (row[4:11])))
            plt.plot(year,male, label='Male')
            plt.plot(year,female, label='Female',  color='olive')
            plt.xlabel('Year')
            plt.ylabel('Average wage')
            if province != 'All Island':
                plt.title( province +" Province  " +" Sector - " +sector)
            if province == 'All Island':
                plt.title( province +" Sector -" +sector)
            plt.legend()
            plt.xticks(year, year)
            plt.show()
        i+=1
     

```
