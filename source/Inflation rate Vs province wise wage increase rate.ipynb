```python
#National inflation rate vs daily average wage increase rate - all island sector wise analysis
import matplotlib.pyplot as plt
import numpy as np
import csv

with open('resources/increase_rate_average_daily_wages.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    i = 0
    plt.figure(figsize=(15,10))
    plt.xlabel('Year')
    plt.ylabel('Average Wage Increase Rate')
    plt.title( "Inflation Rate Vs Average Wage Increase Rate")
    for row in plots:
        if i == 0:
            year = list(map(int, (row[4:10])))
        else :
            province = row[0]
            sector = row[1]
            category = row[2]
            gender = row[3]
            rates = row[4:10]
            rates_list = []
            for rate in rates:
                if not rate:
                    rates_list.append(float('nan'))
                else:
                    rates_list.append(float(rate))
                    
            if province == 'All Island':
                if (sector == 'Inflation Rate'):
                    plt.plot(year,rates_list, '--', label = 'Inflation Rate', color='red')
                    plt.legend()
                else :
                    plt.plot(year,rates_list, label = sector + '_'+ category + '_' + gender)
                    plt.legend()
        i+=1
```

```python
#National inflation rate vs daily average wage increase rate - Province wise and sector wise analysis
import matplotlib.pyplot as plt
import numpy as np
import csv

categories = ['Agriculture_Paddy_Male', 'Construction_Carpentry-Master Carpenter _Male', 'Construction_Masonry-Skilled and Unskilled Helper _Male', 'Agriculture_Tea_Male', 'Agriculture_Paddy_Female', 'Agriculture_Coconut_Male', 'Agriculture_Rubber_Male', 'Agriculture_Tea_Female', 'Construction_Masonry-Master Mason _Male', 'Agriculture_Rubber_Female', 'Construction_Carpentry-Skilled and Unskilled Helper_Male']

with open('resources/increase_rate_average_daily_wages.csv','r') as csvfile:
    for cat in categories :
        plots = csv.reader(csvfile, delimiter=',')
        plt.figure(figsize=(15,10))
        plt.title('Sector:' + cat)
        plt.xlabel('Year')
        plt.ylabel('Average Wage Increase Rate')
        i = 0
        for row in plots:
            if i == 0:
                year = list(map(int, (row[4:10])))
            else :
                province = row[0]
                sector = row[1]
                category = row[2]
                gender = row[3]
                rates = row[4:10]
                rates_list = []
                for rate in rates:
                    if not rate:
                        rates_list.append(float('nan'))
                    else:
                        rates_list.append(float(rate))
                key = sector + '_'+ category + '_' + gender
                if key == cat and province != 'All Island': 
                    plt.plot(year,rates_list, label = province)
                    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
                if (sector == 'Inflation Rate'):
                    plt.plot(year,rates_list, '--', label = 'Inflation Rate', color='red')
                    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
            i+=1
        csvfile.seek(0)
```
